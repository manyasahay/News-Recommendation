```python
from langdetect import detect

def preprocess_text(text):
    if isinstance(text, str):
        # Language detection
        try:
            lang = detect(text)
            if lang != 'en':
                return ""  # Return empty string for non-English text
        except:
            return ""  # In case of error, assume non-English and return empty string

        # Tokenization
        tokens = word_tokenize(text)

        # Lowercasing
        tokens = [token.lower() for token in tokens]

        # Remove punctuation
        tokens = [token for token in tokens if token not in string.punctuation]

        # Remove stopwords
        stop_words = set(stopwords.words('english'))
        tokens = [token for token in tokens if token not in stop_words]

        # Lemmatization
        lemmatizer = WordNetLemmatizer()
        tokens = [lemmatizer.lemmatize(token) for token in tokens]

        # Join tokens back into a single string
        preprocessed_text = ' '.join(tokens)

        return preprocessed_text
    else:
        return ""

df['preprocessed_title'] = df['Title'].apply(preprocess_text)
df['preprocessed_summary'] = df['Summary'].apply(preprocess_text)
```

```python
df.head()
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.preprocessing import OneHotEncoder
```

```python
# Define TF-IDF vectorizer for title and summary
tfidf_vectorizer = TfidfVectorizer(max_features=2000, stop_words='english')
title_tfidf = tfidf_vectorizer.fit_transform(df['Title'].values.astype('U'))
summary_tfidf = tfidf_vectorizer.fit_transform(df['Summary'].values.astype('U'))
```

```python

print(df['Date'].unique())
```

```python
print(tfidf_vectorizer.vocabulary_)
```

```python
print(title_tfidf[:5])  # Print the first 5 TF-IDF vectors for the 'Title' column
print(summary_tfidf[:5])  # Print the first 5 TF-IDF vectors for the 'Summary' column
```

```python
from scipy.sparse import hstack

combined_features = hstack([title_tfidf, summary_tfidf])
```

```python
print(combined_features)
```

```python
from sklearn.cluster import KMeans

kmeans = KMeans(n_clusters=5)  # Example: clustering into 5 clusters
cluster_labels = kmeans.fit_predict(combined_features)
```
